***CLEANING RATE :***

```python
def handle_Rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
zx['Rate']=zx['Rate'].apply(handle_Rate)
zx['Rate'].head()
```
